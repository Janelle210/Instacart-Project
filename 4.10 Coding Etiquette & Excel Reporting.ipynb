```python
midwest_states= ['Wisconsin', 'Michigan' , 'Illinois' ,'Indiana' , 'Ohio','North Dakota', 'South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri']
```

```python
# creating new results for region segmentation
result = []

for value in ords_prods_custs['STATE']:
  if value in northeast_states:
    result.append("Northeast")
  elif value in midwest_states:
    result.append("Midwest")
  elif value in south_states:
    result.append("South")
  elif value in west_states:
    result.append("West")
```

```python
# creating new profiling variables
# setting definition for if-then statement
def profile_variable(row):

  if (row['dependants'] == 0) and (row['Age'] <= 34) and (row['income'] <= 60000):
    return 'no kids, young, low income'
  elif (row['dependants'] == 0) and (row['Age'] >= 35) and (row ['Age'] <=64) and (row['income'] <= 60000):
    return 'no kids, mid, low income'
  elif (row['dependants'] == 0) and (row['Age'] >=65) and (row['income'] <= 60000):
    return 'no kids, old, low income'
  elif  (row['dependants'] == 0) and (row['Age'] <= 34) and (row['income'] > 60000) and (row['income'] < 125000):
    return 'no kids, young, mid income'
  elif (row['dependants'] == 0) and (row['Age'] >= 35) and (row ['Age'] <=64) and (row['income'] > 60000) and (row['income'] < 125000):
    return 'no kids, mid, mid income'
  elif (row['dependants'] == 0) and (row['Age'] >=65) and (row['income'] > 60000) and (row['income'] < 125000):
    return 'no kids, old, mid income'
  elif (row['dependants'] == 0) and (row['Age'] <= 34) and (row['income'] >= 125000):
    return 'no kids, young, high income'
  elif (row['dependants'] == 0) and (row['Age'] >= 35) and (row ['Age'] <=64) and (row['income'] >= 125000):
    return 'no kids, mid, high income'
  elif (row['dependants'] == 0) and (row['Age'] >=65) and (row['income'] >= 125000):
    return 'no kids, old, high income'   
  elif (row['dependants'] >= 1) and (row['Age'] <= 34) and (row['income'] <= 60000):
    return 'parent, young, low income'
  elif (row['dependants'] >= 1) and (row['Age'] >= 35) and (row ['Age'] <=64) and (row['income'] <= 60000):
    return 'parent, mid, low income'
  elif (row['dependants'] >= 1) and (row['Age'] >=65) and (row['income'] <= 60000):
    return 'parent, old, low income'
  elif  (row['dependants'] >= 1) and (row['Age'] <= 34) and (row['income'] > 60000) and (row['income'] < 125000):
    return 'parent, young, mid income'
  elif (row['dependants'] >= 1) and (row['Age'] >= 35) and (row ['Age'] <=64) and (row['income'] > 60000) and (row['income'] < 125000):
    return 'parent, mid, mid income'
  elif (row['dependants'] >= 1) and (row['Age'] >=65) and (row['income'] > 60000) and (row['income'] < 125000):
    return 'parent, old, mid income'
  elif (row['dependants'] >= 1) and (row['Age'] <= 34) and (row['income'] >= 125000):
    return 'parent, young, high income'
  elif (row['dependants'] >= 1) and (row['Age'] >= 35) and (row ['Age'] <=64) and (row['income'] >= 125000):
    return 'parent, mid, high income'
  elif (row['dependants'] >= 1) and (row['Age'] >=65) and (row['income'] >= 125000):
    return 'parent, old, high income'   
```

```python
# creating profile column and where to apply the results. Axis = 1 is all rows, axis = 0 is all columns
ords_prods_custs_clean['profile'] = ords_prods_custs_clean.apply(profile_variable, axis=1)
```
