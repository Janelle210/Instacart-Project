# 03. Grouping Data

```python
# creating a subset of ords_prods_merge
df = ords_prods_merge[:1000000]
```

```python
# checking shape
df.shape
```

```python
# checking column names
df.head()
```

```python
# creating a group by product_name
df.groupby('product_name')
```

# 04. Aggregating Data

```python
# grouping by department id and then aggregating by order number
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# same results, different syntax without "agg"
df.groupby('department_id')['order_number'].mean()
```

```python
# performing multiple aggregations 
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

# 05 Aggregating Data with transform()

```python
# split data into 'user_id' group, 
# apply transform to 'order_number' to generate max orders for each user, 
# and create new column 'max_order'

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

# 06. Deriving columns with loc()

```python
# setting the flags for loyal cusotmers, regular customers and new customers

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# checking the value counts of loyalty flag
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# limiting output of head function to check columns using df['column']
ords_prods_merge[['user_id', 'loyalty_flag', 'max_order']].head(60)
```
